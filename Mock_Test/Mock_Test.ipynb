```
df=spark.read.format('csv').option('header','true').option('inferSchema','true').load('dbfs:/FileStore/comeback/cars.csv')
display(df)
df.count()
```

```
df=df.na.drop()
df.count()
```

```
from pyspark.sql.functions import *
df1=df.withColumn('new_name',lower(df.name))
df1.show()
```

```
df1=df1.filter(df1.new_name.like('%sport%') | df1.new_name.like('%bmw%') | df1.new_name.like('%xuv%') | df1.new_name.like('%jaguar%') | df1.new_name.like('%tata%'))

df1.show()
```

```
df1=df1.withColumn('new_engine',substring(df1.engine,0,4).cast('Int'))

df1.show()
df1.count()
```

```
df1=df1.filter(df1.new_engine>1500).sort(df1.new_engine.desc())
```
